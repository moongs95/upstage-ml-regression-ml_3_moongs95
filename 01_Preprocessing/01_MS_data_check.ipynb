```python
# 좌표X가 NaN인 행만 선택
x_nan_rows = train[train['좌표X'].isna()]

# 확인
x_nan_rows
```

```python
# 결측치 비율 확인 
missing_ratio = train.isnull().mean() * 100
print(missing_ratio.sort_values(ascending=False))
```

```python
# 결측치 개수 계산
missing_counts = train.isnull().sum()
missing_counts
```

```python
# 결측치 70% 이상 컬럼 선택
threshold = 70
cols_to_drop = [col for col in missing_ratio[missing_ratio >= threshold].index if col in train.columns]

exception_cols = ['좌표X', '좌표Y']

# 예외 컬럼 제거
cols_to_drop = [col for col in cols_to_drop if col not in exception_cols]

# 주소/거래 관련 컬럼 추가
cols_to_drop += ['본번', '부번', '등기신청일자', '거래유형', '중개사소재지']

print("삭제할 컬럼들:", cols_to_drop)

# 컬럼 삭제만 수행 (행 삭제는 하지 않음)
train_clean = train.drop(columns=cols_to_drop)

train_clean
```

```python
# 주소 컬럼 만들기 
train_clean['주소'] = train_clean['시군구']+ ' ' + train_clean['번지']
train_clean.head()
```

```python
# 아파트명 결측치 확인
missing_apartment = train_clean[train_clean['아파트명'].isnull()]

# 결측치 있는 행 출력
print(missing_apartment)
# 몇 개 있는지 확인
print("결측치 개수:", missing_apartment.shape[0])
# 아파트명 결측치가 있는 행만 삭제
train_clean = train_clean.dropna(subset=['아파트명'])
train_clean
```

```python
# 아파트 전체 층수가 나와있지 않기 때문에 서로 다른 아파트(최고층이 다 다름)에서 전체 층수를 비교한다는 것이 의미가 없다고 판단
# 아파트 층수에 대한 정보 사용 불가
train_clean = train_clean[train_clean['층'] > 0]
print(train_clean.shape) 
```

```python
# '아파트명' 컬럼에서 '807-39'가 포함된 행 찾기 - 이상한 아파트명 발견
rows = train_clean[train_clean['아파트명'] == '807-39']
print(rows)
```

```python
# train_clean 복사
df_train = train_clean.copy()

# 2. 시군구에서 시를 제외하고 구와 동으로 분리
df_train['구'] = df_train['시군구'].str.split().str[1].reset_index(drop=True)

# 결과 확인
df_train
```
